# **MatplotLib**

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Topics covered
1. Scatter plot
2. Subplots
3. Line
4. Sine and Cosine curve
5. Bar Graph
6. Histogram
7. Boxplot
8. Pie Chart

```python
# Importing matplotlib

import matplotlib.pyplot as plt

# We use the inline command so that we dont have to type plt.show() every time after plotting 

%matplotlib inline
```

```python
import numpy as np
```

```python
# Simple example

x = np.arange(0,10)
y = np.arange(11,21)
```

### **Scatter plot**

```python
# Plotting a scatter diagram using matplotlib 

plt.scatter(x, y, c = 'g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')

# Saving the diagram as a .png image so that we can share it with anyone

plt.savefig("plot.png")
```

```python
# Lets suppose

y = x*x
```

```python
# Plotting

plt.plot(x, y, 'r*', linestyle = 'dashed', linewidth = 2, markersize = 12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Diagram')
```

### **Subplots**

```python
plt.subplot(2,2,1)
plt.plot(x, y, 'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*--')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')

plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

## **Line**

```python
x = np.arange(1,11) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

### **Sine and Cosine curve**

```python
np.pi
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show();
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show();
```

### **Bar Graph**

```python
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

### **Histograms**

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

### **Box Plot**

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot

plt.boxplot(data,vert=True,patch_artist=True);
```

```python
# Lets see the data

data
```

### **Pie Chart**

```python
# Data to plot
labels = 'Python', 'SQL', 'Power Bi', 'Excel'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
